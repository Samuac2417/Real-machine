```python

# Definimos la ruta al archivo CSV (Versión Corregida)
# Subimos solo UN nivel (de 'mlops_pipeline' a 'Proyecto')
data_path = '../Base_de_datos.csv' 

# Cargamos los datos con manejo de errores
try:
    df = pd.read_csv(data_path)
    print(f"Datos cargados exitosamente. Dimensiones: {df.shape}")
    
    # Mostramos las primeras 5 filas para verificar
    display(df.head())

except FileNotFoundError:
    print(f"ERROR: No se pudo encontrar el archivo en la ruta esperada:")
    print(f"Ruta calculada (corregida): {data_path}")
    print("Por favor, verifica que 'Base_de_datos.csv' esté en la carpeta raíz del proyecto ('Proyecto/').")
except Exception as e:
    print(f"Ocurrió un error inesperado al cargar los datos: {e}")
```

```python
# Celda 3: Vistazo General de Tipos de Datos y Nulos
print("--- Información General del DataFrame ---")
df.info()
```

```python
# Celda 4: Estadísticas Descriptivas (Columnas Numéricas)
print("--- Estadísticas Descriptivas (Numéricas) ---")

# display() es una función de Jupyter que formatea la tabla de forma más bonita
display(df.describe())
```

```python
# Celda 5: Reporte de Valores Nulos (en Porcentaje)
print("--- Reporte de Valores Nulos (Columnas con > 0% nulos) ---")

# 1. Calcula el porcentaje de nulos por columna
missing_data = df.isnull().sum() / len(df) * 100

# 2. Filtra solo las columnas que SÍ tienen valores nulos
missing_data = missing_data[missing_data > 0]

# 3. Ordena de mayor a menor para ver las peores
missing_data = missing_data.sort_values(ascending=False)

# 4. Muestra los resultados
if missing_data.empty:
    print("¡Felicidades! No hay valores nulos en el dataset.")
else:
    print("Porcentaje de valores nulos por columna:")
    display(missing_data)
    
    # 5. Opcional pero recomendado: Graficar los nulos
    # Esto nos da un impacto visual inmediato de las columnas más problemáticas
    plt.figure(figsize=(12, 6))
    sns.barplot(x=missing_data.index, y=missing_data.values)
    plt.title('Porcentaje de Valores Nulos por Columna (> 0%)')
    plt.xticks(rotation=90) # Gira las etiquetas para que sean legibles
    plt.ylabel('Porcentaje (%)')
    plt.show()
```

```python
# Celda 6: Análisis de la Variable Objetivo (SalePrice)
print("--- Análisis de la Variable Objetivo: SalePrice ---")

# 1. Calcular la asimetría (skewness). Un valor > 1 es altamente sesgado.
skewness = df['SalePrice'].skew()
print(f"La asimetría (skewness) de SalePrice es: {skewness:.2f}")

# 2. Configurar la visualización (2 gráficos, uno al lado del otro)
fig, axes = plt.subplots(1, 2, figsize=(16, 6)) # 1 fila, 2 columnas

# 3. Gráfico 1: Histograma (para ver la forma de la distribución)
sns.histplot(df['SalePrice'], kde=True, ax=axes[0])
axes[0].set_title('Histograma de SalePrice')
# Añadimos líneas para la media y la mediana
axes[0].axvline(df['SalePrice'].mean(), color='red', linestyle='--', label=f'Media: ${df["SalePrice"].mean():,.0f}')
axes[0].axvline(df['SalePrice'].median(), color='green', linestyle='-', label=f'Mediana: ${df["SalePrice"].median():,.0f}')
axes[0].legend()

# 4. Gráfico 2: Boxplot (para ver los valores atípicos/outliers)
sns.boxplot(x=df['SalePrice'], ax=axes[1])
axes[1].set_title('Boxplot de SalePrice')

plt.suptitle('Distribución del Precio de Venta (SalePrice)', fontsize=16)
plt.show()
```

```python
# Celda 7: Matriz de Correlación (Heatmap de las Top 10 Features)
print("--- Correlación de variables numéricas con SalePrice ---")

# 1. Calculamos la matriz de correlación completa (solo con columnas numéricas)
corr_matrix = df.corr(numeric_only=True)

# 2. Nos quedamos solo con la columna 'SalePrice' y la ordenamos
sale_price_corr = corr_matrix['SalePrice'].sort_values(ascending=False)

# 3. Seleccionamos las 10 características con mayor correlación (incluyendo SalePrice misma)
top_10_features = sale_price_corr.head(10).index
print("Top 10 features más correlacionadas con SalePrice:")
print(sale_price_corr.head(10))

# 4. Creamos una matriz de correlación pequeña, solo con esas 10 features
top_10_corr_matrix = df[top_10_features].corr(numeric_only=True)

# 5. Graficamos el heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(
    top_10_corr_matrix, 
    annot=True,     # Muestra los números de correlación en cada celda
    cmap="coolwarm", # Un mapa de color intuitivo (rojo=positivo, azul=negativo)
    fmt=".2f"       # Formatea los números a 2 decimales
)
plt.title('Heatmap de Correlación (Top 10 Features vs SalePrice)')
plt.show()
```

```python
# Celda 8: Scatterplot de la Característica Top (GrLivArea vs SalePrice)

print("--- Visualizando la relación: GrLivArea vs SalePrice ---")

# Usamos regplot (Regression Plot) porque dibuja el scatter y una línea de tendencia
plt.figure(figsize=(10, 6))
sns.regplot(
    x=df['GrLivArea'], 
    y=df['SalePrice'],
    scatter_kws={'alpha':0.3}, # Pone los puntos semitransparentes para ver la densidad
    line_kws={'color':'red', 'linestyle':'--'} # Dibuja la línea de regresión
)

plt.title('Relación entre Área de Vivienda y Precio de Venta', fontsize=16)
plt.xlabel('GrLivArea (Área de Vivienda en pies cuadrados)')
plt.ylabel('SalePrice (Precio de Venta $)')
plt.grid(True)
plt.show()
```

```python
# Celda 9: Eliminación de Outliers (Valores Atípicos)

# 1. Imprimimos las dimensiones ANTES de eliminar
print(f"Dimensiones originales del dataset: {df.shape}")

# 2. Identificamos y eliminamos los outliers
# Filtramos el DataFrame, quedándonos solo con las filas que NO cumplen esta condición
df = df.drop(df[(df['GrLivArea'] > 4000) & (df['SalePrice'] < 300000)].index)

# 3. Imprimimos las dimensiones DESPUÉS de eliminar
print(f"Dimensiones después de eliminar outliers: {df.shape}")
```

```python
# Celda 10: Verificar el gráfico sin outliers

print("--- Gráfico de GrLivArea vs SalePrice (Sin Outliers) ---")

plt.figure(figsize=(10, 6))
sns.regplot(
    x=df['GrLivArea'], 
    y=df['SalePrice'],
    scatter_kws={'alpha':0.3},
    line_kws={'color':'red', 'linestyle':'--'}
)
plt.title('Relación (Limpia) entre Área de Vivienda y Precio de Venta', fontsize=16)
plt.xlabel('GrLivArea (Área de Vivienda en pies cuadrados)')
plt.ylabel('SalePrice (Precio de Venta $)')
plt.grid(True)
plt.show()
```

```python
# Celda 11: Relación Categórica Clave (OverallQual vs SalePrice)

print("--- Relación entre Calidad General y Precio de Venta ---")

# Creamos un boxplot. Es la mejor forma de ver una variable
# numérica (SalePrice) contra una categórica (OverallQual).
plt.figure(figsize=(12, 7))
sns.boxplot(
    x=df['OverallQual'], 
    y=df['SalePrice']
)
plt.title('Impacto de la Calidad General (OverallQual) en el Precio de Venta', fontsize=16)
plt.xlabel('Calidad General (de 1 a 10)')
plt.ylabel('Precio de Venta ($)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

# Celda 12: Conclusiones del Sprint 1 (EDA)

Basado en el análisis exploratorio de datos, hemos llegado a las siguientes conclusiones clave que guiarán el Sprint 2 (Ingeniería de Características):

1.  **Valores Nulos (Missing Values):**
    * Existen muchas columnas con un alto porcentaje de nulos (ej. `PoolQC`, `Alley`, `Fence`). Estos no son errores, sino que indican **ausencia** de una característica. Deberán ser imputados con el valor "None".
    * Otras columnas (`LotFrontage`, `GarageYrBlt`) tienen nulos que sí son datos faltantes y requerirán una estrategia de imputación más inteligente (ej. rellenar con la mediana o la media).

2.  **Variable Objetivo (`SalePrice`):**
    * La variable `SalePrice` está **fuertemente sesgada a la derecha** (Skewness: 1.88).
    * Para que los modelos de regresión lineal funcionen correctamente, deberemos aplicar una **transformación logarítmica** (ej. `np.log1p`) para normalizarla.

3.  **Características Principales (Top Features):**
    * **Numéricas:** Las variables más correlacionadas con el precio son `GrLivArea` (Área) y `TotalBsmtSF` (Área del Sótano).
    * **Categóricas:** La variable más importante es, por lejos, `OverallQual` (Calidad General).

4.  **Valores Atípicos (Outliers):**
    * Identificamos y **eliminamos 2 outliers** en `GrLivArea` que habrían perjudicado al modelo.
